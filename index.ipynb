```python
# Let's get started importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
# Loading the data
df = pd.read_csv('ames_train.csv')
df.head()
```

```python
# Investigate the Data
df.info()
```

```python
# Investigating Distributions using scatter_matrix
pd.plotting.scatter_matrix(df[['SalePrice','LotArea','YearRemodAdd','YearBuilt']], grid = True, figsize = (10,10))
```

```python
# Create a plot that shows the SalesPrice Distribution
sales_data = df['SalePrice'].plot.hist(bins = 40, figsize=(10,10),histtype='stepfilled')
sales_data.set_xlabel('Sale Price')
sales_data.set_ylabel('Number of Houses')
sales_data.set_title('Sale Price Distribution')
sales_data.axvline(df['SalePrice'].mean(), color='black', linestyle='dashed', linewidth=1)
```

```python
# Create a plot that shows the LotArea Distribution
lot_data = df['LotArea'].plot.hist(bins = 40, figsize=(10,10),histtype='stepfilled')
lot_data.set_xlabel('Lot Area')
lot_data.set_ylabel('Number of Houses')
lot_data.set_title('Lot Area Distribution')
lot_data.axvline(df['LotArea'].mean(), color='black', linestyle='dashed', linewidth=1)
```

```python
# Create a plot that shows the Distribution of the overall house condition
condition_data = df['OverallCond'].plot.hist(figsize=(10,10))
```

```python
# Create a Box Plot for SalePrice
saleprice = df['SalePrice'].plot.box()
```

```python
# Perform an Exploration of home values by age
df['Age'] = 2020 - df['YearBuilt']
home_values = df.plot('Age','SalePrice',figsize = (10,10), kind ='scatter')
```
